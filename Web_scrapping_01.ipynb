```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python'
```

```python
response = requests.get(url)
```

```python
response
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    question = soup.find('a', class_='question-hyperlink')
    print('Question Title : ', question.text)
    
    answers_div = soup.find('div', id='answers')
    answers = answers_div.find_all('div', class_='answer')
    
    for i, answer in enumerate(answers):
        print(f'Answer no. {i + 1} : \n')
        answer_layout = answer.find_all('div', class_='post-layout')
        answer_cell = answer.find('pre', class_=None)
        print(answer_cell.text)
```
